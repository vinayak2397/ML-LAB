```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Load the diabetes dataset
diabetes = load_diabetes()
```

```python
# Use only one feature (age)
X = diabetes.data[:, 2].reshape(-1, 1)
y = diabetes.target
```

```python
# Create a linear regression model and fit it to the data
model = LinearRegression()
model.fit(X, y)
```

```python
# Plot the data points and the line of best fit
plt.scatter(X, y, alpha=0.7)
plt.plot(X, model.predict(X), color='red')
plt.xlabel("Age")
plt.ylabel("Disease progression")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("advertising.csv")

# Select the feature (advertising budget) and target (sales) variables
X = data["TV"].values.reshape(-1, 1)
y = data["Sales"].values


X_mean = np.mean(X)
X_std = np.std(X)
X = (X-X_mean)/X_std


# Add a column of ones to X for the intercept term
X = np.c_[np.ones(X.shape[0]), X]

# Initialize the coefficients to zeros
theta = np.zeros(X.shape[1])

# Set the learning rate and number of iterations
alpha = 0.01
num_iterations = 1000

# Loop over the specified number of iterations
for i in range(num_iterations):
    # Calculate the predicted values
    y_pred = X.dot(theta)
    
    # Calculate the error between the predicted values and the true values
    error = y_pred - y
    
    # Update the coefficients using the LMS algorithm
    theta -= alpha * X.T.dot(error) / X.shape[0]
    
# Print the coefficients
print("Coefficients: ", theta)

# Plot the data points and the line of best fit
plt.scatter(X[:, 1], y, alpha=0.7)
plt.plot(X[:, 1], X.dot(theta), color='red')
plt.xlabel("Advertising Budget ($)")
plt.ylabel("Sales")
plt.show()
```
